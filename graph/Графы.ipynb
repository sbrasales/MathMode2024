При помощи библиотеки NetworkX создайте ненаправленный граф, который состоит из 5 вершин и 4 ребер. Выведите такой граф на экран.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G1 = nx.Graph()      # no dirijido 
G1.add_edges_from([(1, 2), (1, 4), (2, 3), (3, 4), (4, 5)])

nx.draw(G1, with_labels=1)    # muestra los numeros de los nodos
plt.show()
```
