Задача 5

Для всех предыдущих задач вычислите среднее, дисперсию и 0.25 и 0.9 квантили.

Для всех предыдущих задач постройте графики cdf и pdf.

```python
def calculate_stats(mu, sigma):
  
    dist = norm(loc=mu, scale=sigma)
    mean = dist.mean()
    variance = dist.var()
    quantile_25 = dist.ppf(0.25)
    quantile_90 = dist.ppf(0.9)
    
    print(f"среднее: {mean:.2f}")
    print(f"дисперсия: {variance:.2f}")
    print(f"0.25 квантиль: {quantile_25:.2f}")
    print(f"0.90 квантиль: {quantile_90:.2f}")
```

Задача 1

Случайная величина ошибки взвешивания распределена по нормальному закону с нулевым математическим ожиданием и стандартным отклонением 4 грамма. Найти вероятность того, что очередное взвешивание будет проведено с ошибкой, не превышающей по модулю 6 граммов

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

mu1 = 0
std = 4

x = np.linspace(mu1 - 3 * std, mu1 + 3 * std, 1000)

f = norm(mu1, std).pdf(x)
F = norm(mu1, std).cdf(x)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(x, f)
ax[0].grid()

ax[1].plot(x, F)

ax[1].grid()
 
lower_bound = -6
upper_bound = 6
probability = norm.cdf(upper_bound, loc=mu1, scale=std) - norm.cdf(lower_bound, loc=mu1, scale=std)

print(f"probability: {probability:.4f}")

plt.tight_layout()
plt.show()
```

```python
calculate_stats(mu1, std)
```

Задача 2

Автобусы некоторого маршрута идут строго по расписанию и интервалом 5 минут. Составьте функцию плотности случайной величины X – времени ожидании очередного автобуса пассажиром, который наугад подошёл к остановке. Найдите вероятность того, что он будет ждать автобус не более 2 минут.

```python
from scipy.stats import uniform

a = 0  
b = 5  

x = np.linspace(-1, 6, 1000)

f = uniform(a, b-a).pdf(x)  
F = uniform(a, b-a).cdf(x)  

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(x, f)
ax[0].grid()

ax[1].plot(x, F)
ax[1].grid()
prob_2_min = uniform(a, b-a).cdf(2)
print(f"Probability 2 minuts: {prob_2_min:.4f}")

plt.tight_layout()
plt.show()
```

```python
dist_uniform = uniform(a, b - a)
media = dist_uniform.mean()
varianza = dist_uniform.var()
cuantil_25 = dist_uniform.ppf(0.25)
cuantil_90 = dist_uniform.ppf(0.9)

print(f"среднее: {media}")
print(f"дисперсия: {varianza}")
print(f"0.25 квантиль: {cuantil_25}")
print(f"0.90 квантиль: {cuantil_90}")
```

Задача 3

Рост мальчиков возрастной группы 15 лет есть нормально распределённая случайная величина X с математическим ожиданием роста 161 см и среднеквадратичным отклонением 4 см. Какую долю костюмов для мальчиков, имеющих рост от 152 до 158 см, нужно предусмотреть в объёме производства для данной возрастной группы?

```python
mu2 = 161  
sigma2 = 4  

altura_min = 152
altura_max = 158

probabilidad_min = norm.cdf(altura_min, loc=mu2, scale=sigma2)
probabilidad_max = norm.cdf(altura_max, loc=mu2, scale=sigma2)

proportion = probabilidad_max - probabilidad_min

print(f"Proportion between {altura_min} cm y {altura_max} cm is: {proportion:.4f}")

x = np.linspace(mu2 - 4 * sigma2, mu2 + 4 * sigma2, 1000)  
pdf = norm.pdf(x, loc=mu2, scale=sigma2)  
cdf = norm.cdf(x, loc=mu2, scale=sigma2)  

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(x, pdf)
ax[0].grid()

ax[1].plot(x, cdf)

ax[1].grid()
plt.tight_layout()
plt.show()
```

```python
calculate_stats(mu2, sigma2)
```

Задача 4

На машиностроительном заводе в процессе производства за неделю обнаруживается в среднем 100 бракованных деталей со среднеквадричным отклонением в 12 деталей. Рассчитайте следующее: 1) вероятность того, что за неделю будет 20-30 бракованных деталей, 2) вероятность того, что за день будет хотя бы одна забракованная деталь.

```python
mu = 100
sigma= 12

range_min = 20
range_max = 30 

broken = norm(loc=mu, scale = sigma)
min = broken.cdf(range_min)
max = broken.cdf(range_max)

result = max - min
print(f'Defects per week: {result}')


daily_defects = 1 - np.exp(-(mu/7))
print(f'Defects per day: {daily_defects}')

x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)  
pdf = norm.pdf(x, loc=mu, scale=sigma)  
cdf = norm.cdf(x, loc=mu, scale=sigma)  

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(x, pdf)
ax[0].grid()

ax[1].plot(x, cdf)

ax[1].grid()
plt.tight_layout()
plt.show()
```

```python
calculate_stats(mu, sigma)
```
